```python
from graphviz import Digraph
# Create Digraph object
dot = Digraph()
# Add nodes 1 and 2
dot.node('A1','ADD')
dot.node('S1','SUB')

# Add edge between 1 and 2
dot.edge('A1', 'S1')
# Visualize the graph
dot
```
